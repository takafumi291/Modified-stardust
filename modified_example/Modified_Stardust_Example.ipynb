### Example: Stardust fit using point source flux as AGN emission

This code reproduces the result in Tsukui et al. 2023

```python
import stardust
import numpy as np
```

Load in the class by proving the path to the .config file.

```python
self = stardust.main.ctf(config_file='bri1335_agn_enabled.config')
```

We can double check what parameters are in the config file

```python
self.config
#To use modified stardust, you only need to set these parameters:
#AGN_EXTRA_BANDS_FILE
#AGN_EXTRA_BANDS = 1
```

These can also be changed manually after loading the module e.g.

```python
self.config['OUTPUT_NAME'] = 'bri1335'
```
